## PANDAS - parte III

```python
import pandas as pd
import numpy as np

df = pd.read_csv('alunos2017.csv')
```

```python
from IPython.display import Image
Image(url='http://guiadocorpo.com/wp-content/uploads/2015/08/Formula-do-IMC.png')
```

```python
# Exercicio 1: calcular o IMC

df['IMC'] = df.Peso/((df.Altura/100) ** 2)
```

```python
Image(url='https://viveresano.files.wordpress.com/2010/07/imc.jpg')
```

```python
# Exercicio 2: transformar o IMC em uma variavel categórica

df['graus_de_obesidade'] = pd.cut(df.IMC, [0, 18.5, 25, 30, 35, 40, np.inf],
                                 labels = ['baixo', 'normal', 'pre-obeso',
                                          'grau I', 'grau II', 'grau III'])
```

```python
# apagar graus_de_obesidade
df.drop('graus_de_obesidade', axis = 1, inplace = True)
```

## Groupby

```python
df.head()
```

```python
df.groupby('Time')
```

```python
for time, grupo in df.groupby('Time'):
    print (time)
    print (grupo)
    print ('-'* 30)
```

```python
# definir um objeto agrupado por Time
df_agrupado = df.groupby('Time')
```

```python
# quais são os menores IMC por time?
df_agrupado.IMC.min()
```

```python
# quais são as medias das colunas por time?
df_agrupado.mean()
```

```python
# função que retorna a diferença entre max e min:
def dif_max_min(arr):
    return arr.max() - arr.min()

# teste
dif_max_min(df.IMC)
```

```python
# agg: passar como argumento uma função a ser aplicada sobre cada grupo
df_agrupado.agg(dif_max_min)
```

```python
# algumas funções já existem e só passar o str como argumento
df_agrupado.agg('mean')
```

```python
# criar uma coluna nova com um ranking de IMC

df.sort_values('IMC', inplace=True)
def ranker(df):
    df['ranking'] = np.arange(len(df)) + 1
    return df
```

```python
df_agrupado = df.groupby('Time')
df_rankeado = df_agrupado.apply(ranker)
df_rankeado
```

```python
# para lidar com o times não informados
df['Time'].fillna('ND', inplace = True)
df.sort_values('IMC', inplace=True)
```

```python
df_agrupado = df.groupby('Time')
df_rankeado = df_agrupado.apply(ranker)
df_rankeado
```

## concat e merge

```python
df_pal = df[df.Time == 'Palmeiras']
df_nan = df[df.Time == 'ND']
```

```python
df_pal
```

```python
df_nan
```

```python
# concatenar os dois dataframes df_pal e df_nan
df_concat = pd.concat([df_pal, df_nan])
df_concat
```

```python
# criar df_flu e concatenar
df_flu = df[df.Time == 'Fluminense']
df_concat = pd.concat([df_concat, df_flu])
```

```python
# criando um dataframe a partir de um dict
df_uf = pd.DataFrame({'Time': ['Palmeiras', 'ND', 'Fluminense'], 'UF': ['SP', 'BR', 'RJ']})
df_uf
```

```python
# merge df_concat com df_uf
pd.merge(df_concat, df_uf)
```
