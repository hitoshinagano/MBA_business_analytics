## checando os tipos

diferenças entre type(), .dtype() e o tipo dos elementos:

* `type()` retorna o tipo do objeto, sendo um comando genérico em python refere-se a classe definida pelo módulo pandas, tipicamente `Series` ou `DataFrame`
* `.dtype` é um atributo do objeto, representando os tipos de dados contidos, tipicamente em linhas e colunas
* cada elemento contido dentro de uma série ou dataframe tem seu tipo que pode ser obtido por `type()`.

veja os exemplos a seguir

````python
print(type(df.PTS))
print(df.PTS.dtype)
print(type(df.PTS.iloc[1]))
````

````python
print(type(df.Equipe))
print(df.Equipe.dtype)
print(type(df.Equipe.iloc[1]))
````

repare que o *dtype* `object` é como se fosse um "coringa", assim designado quando pelo menos um dos elementos da série é um string. Genericamente, somente quando todos os elementos da série aderem a um mesmo tipo numérico, é que o dtype correspondente é também numérico.

## indexação e fatiamento (*indexing* e *slicing*)

é possível fazê-lo por algumas maneiras distintas:

* pelo nome da coluna (ou linha): retorna uma série pandas
* `.loc[]`: indexa/fatia pelo nome do indice e/ou coluna  (\*)
* `.iloc[]`: indexa/fatia por um número inteiro sequencial começando do `0` (\*)

(\*): retorna uma série (fatiamento) ou o valor (indexação)

````python
# coluna - metodo 1
df['PTS']
````

````python
# coluna - metodo 2
df.PTS
````

````python
df.index
````

````python
# usando .loc : indexação pelo nome linha
df.loc['SP Corinthians']
````

````python
# usando .loc : indexação pelo nome de linha e de coluna
df.loc['SP Corinthians', 'V']
````

````python
# usando .iloc: indexação por um número (como se fosse um indice alternativo e sequencial)
df.iloc[1:4,]
````

## operações

vejamos algumas operações com colunas do *DataFrame*. Note mais uma vez a "sobrecarga de operador" em ação. O que faz mais sentido nesse objeto colunar é realmente somar termo-a-termo.  

````python
# + sobre colunas soma termo a termo
df.V + df.E + df.D
````

para somar, recorremos ao metodo `.sum()`, cujo valor default é `axis = 0`. Ou seja, uma iteração sobre as linhas (cujo resultado é a soma dos valores das colunas).

````python
df.sum()
````

apesar de não fazer sentido nesse contexto, se quisessemos somar a linha temos que passar explicitamente `axis = 1`. Ou seja, uma iteração sobre as colunas (cujo resultado é a soma dos valores das linhas).

````python
df.sum(axis = 1)
````

outros métodos incluem:

* `.mean()`: média
* `.std()`: desvio padrão

````python
df.mean()
````

````python
df.std()
````
